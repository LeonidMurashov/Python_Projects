```python
import json
import requests

posts = []
group_damain = "science_technology"

wall_data = requests.get("https://api.vk.com/method/wall.get?domain=" + group_damain + "&v=5.60&count=100")
posts_response = json.loads(wall_data.text)
count = posts_response["response"]["count"]

for i in range(0, count, 100):
    wall_data = requests.get("https://api.vk.com/method/wall.get?domain=" + group_damain + "&v=5.60&count=100&offset="+str(i))
    posts_response = json.loads(wall_data.text)
    posts = posts + posts_response["response"]["items"]
    print(i)
```

```python
top_posts = []

for post in posts:
    top_posts.append({"id" : post["id"], "likes" : post["likes"]["count"], "reposts" : post["reposts"]["count"], "text" : post["text"]})
        
top_posts = sorted(top_posts, key= lambda x: x["likes"], reverse=True)
```
